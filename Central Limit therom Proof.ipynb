# Step -1 Sampling / Randomly pick samples from the data

```python
import random
```

```python
np.mean(random.choices(dataset["Purchase"],k=20))#dataset,k- no of observations
```

```python
n=100 #total no of samples
k1=100 #total no of observation
```

```python
sample_data=[]
```

```python
for i in range(0,n):
    sample_data.append(random.choices(dataset["Purchase"],k=k1))
```

```python
sample_data
```

```python
sample_means=[]
```

```python
for i in sample_data:
    sample_means.append(np.mean(i))
```

```python
sample_means
```

```python
sns.distplot(sample_means)
```

# Mean of sample means

```python
smean=np.mean(sample_means)
```

```python
#Sample Mean
smean #k=30
```

```python
#Population mean
dataset["Purchase"].mean()
```

# Variance

```python
svar=np.var(sample_means)
```

```python
svar
```

```python
#Population variance
dataset["Purchase"].var()/100
```

# Standard Deviation

```python
sstd=np.std(sample_means)
```

```python
sstd
```

```python
#Population std
dataset["Purchase"].std()/np.sqrt(100)
```

# Hence Proved
1. Sample Statistic will approximate population statistic when sample size high and randomly picked
