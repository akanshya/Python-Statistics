```python
dataset.head()
```

```python
q1=dataset.describe()["Profit"]["25%"]
```

```python
q3=dataset.describe()["Profit"]["75%"]
```

```python
q1
```

```python
q3
```

```python
iqr=q3-q1
```

```python
iqr
```

```python
lower_bound=q1-(1.5*iqr)
```

```python
lower_bound
```

```python
upper_bound=q3+(1.5*iqr)
```

```python
upper_bound
```

# Remove the Outliers

```python
out_index=dataset[(dataset["Profit"]<lower_bound) | (dataset["Profit"]>upper_bound)].index
```

```python
dataset.drop(out_index,inplace=True)
```

```python
sns.boxplot(dataset["Profit"])
```

# Replace the Outliers

```python
out_l=dataset["Profit"][(dataset["Profit"]<lower_bound)].values
```

```python
out_l
```

```python
data
```

# Step-5 Encoding Categorical Data

```python
dataset
```

# OneHotEncoder

```python
dataset=dataset.values
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.compose import make_column_transformer
```

```python
ohe=OneHotEncoder()#Object of OneHotEncoder
```

```python
#(name, transformer, column(s))
#1st element -name Of Transformation
#2nd element -- instance/Object Transformation
#3rd element -- [column]
#remainder --Operation to performed to remaining columns -Drop,passthrough
ct=ColumnTransformer(transformers=[("oh",ohe,[3])],remainder="passthrough")
```

```python
dataset=ct.fit_transform(dataset)
```

```python
dataset=dataset.astype("float")
```

```python
dataset
```
